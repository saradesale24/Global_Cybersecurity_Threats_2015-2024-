```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**This scatter plot shows the relationship between the number of affected users and financial loss for cyber incidents in 2022, where each point represents a single attack; it helps to identify whether attacks affecting more users tend to cause higher financial losses, or if losses vary independently of user count, revealing patterns or outliers in that year’s data.**

```python
#Financial Loss vs Incident Resolution Time for attacks in 2020
subset = df[df['Year'] == 2020]
sns.scatterplot(data=subset, x='Incident Resolution Time (in Hours)', y='Financial Loss (in Million $)', hue='Attack Type')
plt.title('2020 Attacks: Loss vs Resolution Time')
plt.show()
```

**This scatter plot shows the relationship between financial loss and incident resolution time for cyberattacks in 2020, with each point representing an attack; the x-axis shows resolution time, the y-axis shows financial loss, and colors distinguish attack types, helping to observe which types caused higher losses or required longer resolution.**

```python
#Financial Loss vs Affected Users
sns.scatterplot(data=df, x="Number of Affected Users", y="Financial Loss (in Million $)", hue="Attack Type", s=80)
plt.xscale("log"); plt.yscale("log"); plt.title("Loss vs Users"); plt.tight_layout(); plt.show()
```

**This scatter plot shows the relationship between the number of affected users and financial loss across all attack types, using a log scale on both axes to handle wide data ranges; each point is an attack, colored by type, helping to identify whether attacks with larger user impact generally lead to higher financial losses or if the relationship varies by attack type.**

```python
#Affected Users by Defense Mechanism

sns.kdeplot(data=df, x="Number of Affected Users", hue="Defense Mechanism Used", fill=True)
plt.title("KDE of Affected Users by Defense Mechanism")
plt.show()
```

**This KDE plot shows the distribution of affected users across different defense mechanisms. Peaks indicate where incidents were most concentrated, while wider curves show greater spread. Comparing curves reveals which defense mechanisms were linked to higher or lower user impacts.**

```python
#Country-wise Resolution Times

sns.kdeplot(data=df, x="Incident Resolution Time (in Hours)", hue="Country",fill=True)
plt.title("KDE of Resolution Time by Country")
plt.show()
```

**This KDE plot shows the distribution of incident resolution times across different countries. Peaks represent the most common resolution durations in each country, while curve width indicates variability — wider spreads suggest inconsistent resolution times compared to sharper peaks.**

```python
#Yearly Trend of Financial Loss
sns.relplot(data=df, x="Year", y="Financial Loss (in Million $)", hue="Country", kind="line", marker="o")

```

**This line plot shows the yearly trend of financial loss across different countries. Each line represents a country, markers highlight yearly values, and the slope of the line indicates whether losses are increasing, decreasing, or fluctuating over time.**
